```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
from pathlib import Path
```

```python
DATA_PATH=Path('../data/raw/public_transport_delays.csv')
df=pd.read_csv(DATA_PATH)
```

```python
df['date'] = pd.to_datetime(df['date'])
df['event_type'] = df['event_type'].fillna('No events')
```

```python
df['month'] = df['date'].dt.month
df['hour'] = pd.to_datetime(df['time'], format='%H:%M:%S').dt.hour
```

```python
df['scheduled_travel_time_min'] = (
    pd.to_datetime(df['scheduled_arrival'], format='%H:%M:%S') -
    pd.to_datetime(df['scheduled_departure'], format='%H:%M:%S')
).dt.seconds / 60
```

```python

print(df['scheduled_travel_time_min'].min())
print((df['scheduled_travel_time_min'] < 0).sum())
```

```python
cols_to_drop = [
    'trip_id', 'date', 'time',
    'scheduled_departure', 'scheduled_arrival',
    'actual_departure_delay_min',
    'route_id', 'origin_station', 'destination_station', 'month'
]
df = df.drop(columns=cols_to_drop)
```

```python
#df.select_dtypes(include='object')
df.select_dtypes(include='str')
```

```python
cat_cols=['transport_type','weather_condition','event_type','season']
df = pd.get_dummies(df, columns=cat_cols, drop_first=True)
```

```python
df.to_csv('../data/processed/transport_processed.csv', index=False)
```
